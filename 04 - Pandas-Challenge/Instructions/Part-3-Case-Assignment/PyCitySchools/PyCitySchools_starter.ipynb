```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete
```

```python
#Calculate the total number of schools
Total_num_schools = school_data_complete["school_name"].nunique()

#Calculate the total number of students
Total_num_students = school_data_complete["student_name"].count()

#Calculate the total budget
budget = school_data_complete["budget"].unique()
Total_budget = round((budget.sum()),2)

# Calculate the Average Scores
avg_math_score = school_data_complete["math_score"].mean()

avg_reading_score = school_data_complete["reading_score"].mean()

# Calculate the Percentage Pass Rates
mathpassedcount = school_data_complete["math_score"].loc[school_data_complete["math_score"] >=70].count()
mathpasspercent=(mathpassedcount/Total_num_students)*100

readpassedcount = school_data_complete["reading_score"].loc[school_data_complete["reading_score"] >=70].count()
readpasspercent=(readpassedcount/Total_num_students)*100


#overallpasspercent = (mathpasspercent + readpasspercent)/2
overallpasscount = school_data_complete["student_name"].loc[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"]>=70)].count() 
overallpasspercent =(overallpasscount/Total_num_students)*100


school_summary_df =pd.DataFrame(
        {
            "Total Schools": [Total_num_schools]
            ,"Total Students": [Total_num_students]
            ,"Total Budget" : [Total_budget]
            ,"Average Math Score" : [avg_math_score]
            ,"Average Reading Score" : [avg_reading_score]
            ,"% Passing Math" : [mathpasspercent]
            ,"% Passing Reading" : [readpasspercent]
            ,"% Overall Passing Rate" : [overallpasspercent]
        }
                                )




```

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Determine the School Type
#Create an overview table that summarizes key metrics about each school, including:
school_grp = school_data_complete.groupby(["school_name", "type"])
school_grp1 = school_data_complete.groupby(["school_name", "type"]).mean()
students_per_school=school_grp.count()["student_name"]

per_school_budget = school_grp.mean()["budget"]

per_student_budget = per_school_budget/students_per_school

per_school_avgm = school_grp.mean()["math_score"]

per_school_avgr = school_grp.mean()["reading_score"]

math_pass_percent = school_data_complete[(school_data_complete["math_score"]>=70)]
math_pass_percent = math_pass_percent.groupby(["school_name","type"]).count()["student_name"]/students_per_school*100

read_pass_percent = school_data_complete[(school_data_complete["reading_score"]>=70)]
read_pass_percent = read_pass_percent.groupby(["school_name","type"]).count()["student_name"]/students_per_school*100

#overall_pass_percent = (math_pass_percent + read_pass_percent)/2
overall_pass_percent = school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)]
overall_pass_percent = overall_pass_percent.groupby(["school_name","type"]).count()["student_name"]/students_per_school*100
#overall_pass_percent
```

```python
#school_grp1 
perschool_summary = pd.DataFrame(school_grp1)
perschool_summary["Total Students"] = students_per_school
perschool_summary["Total School Budget"] = per_school_budget
perschool_summary["Per Student Budget"] = per_student_budget
perschool_summary["Average Math Score"] = per_school_avgm
perschool_summary["Average Reading Score"] = per_school_avgr
perschool_summary["% Passing Math"] = math_pass_percent
perschool_summary["% Passing Reading"] = read_pass_percent
perschool_summary["% Overall Passing Rate"] = overall_pass_percent
perschool_summary = perschool_summary.drop(columns=["Student ID", "reading_score","math_score","size","School ID","budget"])

perschool_summary.reset_index(inplace = True)
perschool_summary = perschool_summary.rename(columns={"school_name": "School Name"
                                                      ,"type":"School Type"})
perschool_summary.head(20)
```

```python
# Formatting

perschool_summary_new = perschool_summary.copy()
perschool_summary_new["Total School Budget"] = perschool_summary_new["Total School Budget"].map("${:,}".format)
perschool_summary_new["Per Student Budget"] = perschool_summary_new["Per Student Budget"].map("${:}".format)

perschool_summary_new
```
