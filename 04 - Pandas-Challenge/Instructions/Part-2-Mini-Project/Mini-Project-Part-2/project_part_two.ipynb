```python
# Dependencies
import pandas as pd
```

```python
# load CSV
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv",encoding='ISO-8859-1',low_memory=False)
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
# You may need to encode this file
df.head()
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Consider using iloc

columns = [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
new_df = df.iloc[:,columns]
new_df.head(10)
```

```python
# Change reading 0 and 1 to No and Yes, respectively

new_df["AttendedBootcamp"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["BootcampFinish"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["BootcampFullJobAfter"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["BootcampLoanYesNo"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["BootcampRecommend"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["CodeEventBootcamp"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["HasChildren"].replace({1: "Yes", 0: "No"},inplace=True)

new_df

```

```python
# Extract rows for only those who attended a bootcamp
attend_bootcamp_df = new_df.loc[new_df['AttendedBootcamp'] == 'Yes']
attend_bootcamp_df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
# Create a DataFrame from using value_counts
attendee_count = attend_bootcamp_df['BootcampName'].value_counts() 
attendee_count
attendee_count_df=pd.DataFrame(attendee_count)
attendee_count_df.reset_index(inplace=True)
attendee_count_df.columns=['BootcampName','Attendees_count']
attendee_count_df
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp

r_attend_bootcamp1 =  attend_bootcamp_df.loc[attend_bootcamp_df['BootcampRecommend'] == 'Yes']
#r_attend_bootcamp_df.head()
r_attend_bootcamp_count=r_attend_bootcamp1['BootcampName'].value_counts()
r_attend_bootcamp_count_df=pd.DataFrame(r_attend_bootcamp_count)
r_attend_bootcamp_count_df.reset_index(inplace=True)
r_attend_bootcamp_count_df.columns=['BootcampName','Recommeneded_count']
r_attend_bootcamp_count_df
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_df = pd.merge( attendee_count_df,r_attend_bootcamp_count_df, on="BootcampName",how='right')
merge_df.head()
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
# Think of what should be the numerator and the demoninator
merge_df["% Recommend"] = round((merge_df["Recommeneded_count"]/merge_df["Attendees_count"])*100,2)

# Sort results in descending order
merge_df =merge_df.sort_values("% Recommend",ascending = False)

# Format for percentages
merge_df["% Recommend"] = merge_df["% Recommend"].map("{:}%".format)
merge_df.head()
```

```python
# Export to excel and remove index
merge_df.to_excel("output/summary_output2.xlsx", index=False)
```
