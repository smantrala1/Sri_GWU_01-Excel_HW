```python
# Dependencies
import pandas as pd
import os
```

```python
# load CSV
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv",encoding='ISO-8859-1',low_memory=False)
```

```python
# Read with pandas
df.head()
```

```python
# Inspect all columns
df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)

columns = [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
new_df = df.iloc[:,columns]
new_df.head(10)
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function
new_df["AttendedBootcamp"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["BootcampFinish"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["BootcampFullJobAfter"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["BootcampLoanYesNo"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["BootcampRecommend"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["CodeEventBootcamp"].replace({1: "Yes", 0: "No"},inplace=True) 
new_df["HasChildren"].replace({1: "Yes", 0: "No"},inplace=True)

new_df
```

```python
# Calculate total number of respondents in survey
# consider using the len() function

totnumofResp = len(new_df.index)
print(totnumofResp)
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
bootcamp_df = new_df.loc[new_df['AttendedBootcamp'] == 'Yes']
bootcamp_df.head(10)
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it

avg_age =round(bootcamp_df["Age"].mean(),2)
avg_age
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column
bootcamp_attendence = len(bootcamp_df.index)
print(bootcamp_attendence)
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column

bootcamp_df['SchoolDegree'].value_counts()
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column

# 1 way
#degree_count = bootcamp_df['SchoolDegree'].loc[(bootcamp_df['SchoolDegree'] == "bachelor's degree")  | (bootcamp_df['SchoolDegree'] == "master\'s degree (non-professional)") | (bootcamp_df['SchoolDegree'] == "professional degree (MBA, MD, JD, etc.)") | (bootcamp_df['SchoolDegree'] == "associate's degree") | (bootcamp_df['SchoolDegree'] == "Ph.D.")].count()
#2 - more efficient than above. 
degree_count = bootcamp_df['SchoolDegree'].loc[(bootcamp_df['SchoolDegree'].str.contains("degree") == True) & (bootcamp_df['SchoolDegree'].str.contains("no degree") == False) |  (bootcamp_df['SchoolDegree'] == "Ph.D.")].count()                      
                   
degree_count
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
Total = bootcamp_df["Gender"].count()
Male_count = bootcamp_df["Gender"].loc[(bootcamp_df['Gender'] == "male")].count()
Female_count = bootcamp_df["Gender"].loc[(bootcamp_df['Gender'] == "female")].count()
non_specific = bootcamp_df["Gender"].loc[(bootcamp_df['Gender'] != "female") & (bootcamp_df['Gender'] != "male")].count()
print(f"Total: {Total}")
print(f"Male: {Male_count}")
print(f"Female: {Female_count}")
print(f"Non-Specific: {non_specific}")
```

```python
# Calculate percentage of respondents who attended a bootcamp
#totnumofResp
Percentage_respondents = round((bootcamp_attendence/totnumofResp)*100,2)
Percentage_respondents
```

```python
# Calculate percentage of respondents belonging to each gender
# % Male: 59.5438175270108
# % Female: 39.13565426170468
# % non_specifc: 11
Male_perc = round((Male_count/Total)*100,2)
Female_perc = round((Female_count/Total)*100,2)
nons_perc = round((non_specific/Total)*100,2)
print(f"Male Percentile: {Male_perc}")
print(f"Female Percentile: {Female_perc}")
print(f"Non-Specific Percentile: {nons_perc}")
```

```python
# Calculate percentage of attendees with a college degree
#66.84155299055614
degree_perc = round((degree_count/bootcamp_attendence)*100,2)
degree_perc
```

```python
# Calculate average post-bootcamp salary

pbc_salary =round(bootcamp_df["BootcampPostSalary"].mean())
pbc_salary
```

```python
# Create a new table consolidating above calculations

summary_df =pd.DataFrame({
    "Total Surveyed": [totnumofResp]
    ,"Total Bootcamp attendees": [bootcamp_attendence]
    ,"% attended bootcamp" : [Percentage_respondents]
    ,"Avg. Age" : [avg_age]
    ,"Has a degree" : [degree_perc]
    ,"% Male" : [Male_perc]
    ,"% Female" : [Female_perc]
    ,"% Non Gender Specific" : [nons_perc]
    ,"Average Post Bootcamp Salary" : [pbc_salary]
})
summary_df
```

```python
# Improve formatting before outputting spreadsheet
summary_df["Average Post Bootcamp Salary"] = summary_df["Average Post Bootcamp Salary"].map("${:,}".format)
summary_df["% attended bootcamp"] = summary_df["% attended bootcamp"].map("{:}%".format)
summary_df["Has a degree"] = summary_df["Has a degree"].map("{:}%".format)
summary_df["% Male"] = summary_df["% Male"].map("{:}%".format)
summary_df["% Female"] = summary_df["% Female"].map("{:}%".format)
summary_df["% Non Gender Specific"] = summary_df["% Non Gender Specific"].map("{:}%".format)


summary_df
```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes

summary_df.to_excel("output/summary_output.xlsx", index=False)
```
